# Let's do the same to all the months

```python
import pandas as pd

def read_and_process_csv(file_path,skiprows=9):
    # Read the CSV file, skipping the first 9 rows and filling NaN values with 0
    df = pd.read_csv(file_path, skiprows=skiprows).fillna(0)
    return df

# File paths for January to December
file_paths = [
    "Chilly Mazarin/Suivi_des_ventes_horaires January 2023 Chilly.csv",
    "Chilly Mazarin/Suivi_des_ventes_horaires March 2023 Chilly.csv",
    "Chilly Mazarin/Suivi_des_ventes_horaires April 2023 Chilly.csv",
    "Chilly Mazarin/Suivi_des_ventes_horaires May 2023 Chilly.csv",
    "Chilly Mazarin/Suivi_des_ventes_horaires June 2023 Chilly.csv",
    "Chilly Mazarin/Suivi_des_ventes_horaires July 2023 Chilly.csv",
    "Chilly Mazarin/Suivi_des_ventes_horaires August 2023 Chilly.csv",
    "Chilly Mazarin/Suivi_des_ventes_horaires September 2023 Chilly.csv",
    "Chilly Mazarin/Suivi_des_ventes_horaires October 2023 Chilly.csv",
    "Chilly Mazarin/Suivi_des_ventes_horaires November 2023 Chilly.csv",
    
]
    

# Create a dictionary to store dataframes for each month
dataframes = {}

# Read and process each CSV file
for file_path in file_paths:
    # Extract month from file path
    month = file_path.split()[2]
    # Read CSV file and assign to dataframe
    dataframes[month] = read_and_process_csv(file_path)
# The dataset are of different format for feburary and december
dataframes['Feburary'] = pd.read_csv("Chilly Mazarin/Suivi_des_ventes_horaires February 2023 Chilly.csv",skiprows=16,sep=';')
dataframes['December'] = pd.read_csv("Chilly Mazarin/Suivi_des_ventes_horaires December 2023 Chilly.csv", skiprows=10, sep=';').fillna(0)

# Now we have dataframes for each month stored in the 'dataframes' dictionary
# Create a new dictionary for transaction data
monthly_dataframe = {}

for month,df in dataframes.items():
    idx =np.where(df['Mesures']=='Transaction')[0]
    monthly_dataframe[month] = df.iloc[idx,:]

#December doesn't have transaction but net sales
monthly_dataframe['December'] = dataframes['December'].iloc[np.where(dataframes['December']['Mesures']=='Net Sales')[0],:]
```

```python
monthly_dataframe['Feburary']
```

# Here we have to make two assumption:
### Assumption1: for each restaurant, its monthly alignment of turnover with kitchen energy consumption is constant
### Assumption2: for each restaurant, its monthly pattern of kitchen energy consumption is constant

Since we only have the turnover of chilly, then we can for now only do the analysis of alignment for this restaurant

```python
# First we take the overall monthly transaction by hour

# Define the start and end hour
start_hour = '00:00'
end_hour = '23:00'

# Generate the list of hour intervals in the format "00:00-00:59"
hourly_intervals = [(pd.to_datetime(hour).strftime('%H:%M') + '-' +
                     (pd.to_datetime(hour) + pd.Timedelta(hours=1) - pd.Timedelta(seconds=1)).strftime('%H:%M'))
                    for hour in pd.date_range(start=start_hour, end=end_hour, freq='h')]

# Create an empty DataFrame with columns representing each hour interval
transaction_chilly_2023 = pd.DataFrame(columns=hourly_intervals)

# Display the empty DataFrame
print(transaction_chilly_2023)
```

```python
for month,df in monthly_dataframe.items():
    common_hours = [hour for hour in hourly_intervals if hour in df.columns]
    for hour in common_hours:
        transaction = df.iloc[0,df.columns.get_loc(hour)]
        transaction_chilly_2023.at[month,hour]= transaction
    
```

```python
transaction_chilly_2023.fillna(0)
```

```python
# tackle with the comma in value
transaction_chilly_2023 = transaction_chilly_2023.apply(lambda x: x.astype(str))
transaction_chilly_2023 = transaction_chilly_2023.apply(lambda x: x.str.replace(',', '').astype(float))
```

```python
plt.figure(figsize=(20, 12))
for i in range(12):
    plt.plot(transaction_chilly_2023.iloc[i,1:], label=transaction_chilly_2023.index[i])
plt.xticks(rotation=45)
plt.xlabel('Hour')
plt.ylabel('Turnover ')
plt.title('Hourly Turnover')
plt.legend()
plt.show()
```
